There are only 103,588 unique values of 'vin'. Some of these values are missing values, let's see how many missing values we have in the vin column

```python
df['vin'].isnull().sum()
```

There're a total of 423,857  rows in the vehicles dataframe, out of which 56.49% (239,437) of vehicles identification numbers (vin) are not entered. This means we should have (423,857 - 239,437) = 184,420 unique vin's since no two vehicles can have the same vin,  but on the contrary, there are only 103,588 unique vin's, which means 80,832 vin's are duplicated. 


We will keep the 239,437 rows with null 'vin' values in a separate dataframe, drop vin duplicates and then concatenate both dataframes. The resulting dataframe will have (239,437 + 103,588)  = 343,025 rows

```python
null_vin = df[df['vin'].isnull()]
null_vin
```

```python
#check for duplicate entries in the vin column after dropping null values
vinduplicated = df['vin'].dropna().duplicated(keep = 'first').value_counts()
vinduplicated
```

```python
vin_df = df[df['vin'].notnull()]
vin_df
```

```python
#Drop duplicated columns from vin_df
vins = vin_df['vin']
vin_unique = vin_df.drop_duplicates(subset = ['vin'])
vin_unique
```

```python
#Concat vin_unique with null_vin
vehicles_df = pd.concat([vin_unique, null_vin])
vehicles_df
```

Now, vehicles with duplicated entries are dropped

```python
vehicles_df.info()
```

```python
#Save the new datafame as csv
vehicles_df.to_csv('UsedVehiclesWithNoDuplicatevin.csv', index=False)
```
